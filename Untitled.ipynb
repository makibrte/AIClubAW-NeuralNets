```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
```

```python
class SimpleNet(nn.Module):
    def __init__(self, in_features=3, hidden_dim = 10, out_dim = 1):
        super().__init__()
        
        self.linear_1 = nn.Linear(in_features, hidden_dim)
        self.linear_2 = nn.Linear(hidden_dim, out_dim)
    def forward(self, x):
        x = F.relu(self.linear_1(x))
        x = self.linear_2(x)
        return x
```

```python
x = torch.tensor([[x] for x in np.arange(0,10, 0.1)], dtype=torch.float)
y = torch.tensor([[np.sin(x)] for x in np.arange(0,10, 0.1)], dtype=torch.float)
```

```python
plt.plot(x, y)
```

```python
model = SimpleNet(1, 100, 1)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-2)
criterion = nn.MSELoss()
```

```python
outputs_list = []
losses = []
for epoch in range(1000):
    # setting the gradients to 0
    optimizer.zero_grad()

    outputs = model(x)
    # compute the loss
    loss = criterion(y, outputs)
    # backward propagation for the loss
    loss.backward()
    losses.append(loss.item())
    # once gradients are passed "step using" optimizer
    optimizer.step()    

    outputs_list.append(outputs)
```

```python
plt.plot(x.detach().numpy(), outputs.detach().numpy())
```
